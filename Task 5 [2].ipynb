```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = 3 * x + 2

plt.plot(x, y, label='y = 3x + 2', color='blue')

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.title('Line Chart of y = 3x + 2')
plt.legend()

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(8, 6))
plt.scatter(
    x=x,
    y=y,
    c='#FF7F50',
    s=100,
    alpha=0.7,
    edgecolor='k',
    linewidths=1,
)

plt.grid(True, linestyle='--', alpha=0.4)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.title('Scatter Plot of 50 Random Points', fontweight='bold')

plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['Apple', 'Samsung', 'Google', 'OnePlus', 'Xiaomi']
values = [45, 38, 12, 15, 25]


colors = ['#2E86C1', '#3498DB', '#5DADE2', '#85C1E9', '#AED6F1']

plt.figure(figsize=(10, 6))
bars = plt.bar(
    categories,
    values,
    color=colors,
    edgecolor='black',
    linewidth=0.7
)

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 1,
        f'{height}%',
        ha='center',
        va='bottom',
        fontweight='bold'
    )

plt.title('Market Share of Mobile Brands (%)', fontsize=14, pad=20)
plt.xlabel('Brand', fontsize=12)
plt.ylabel('Percentage', fontsize=12)
plt.ylim(0, 50)

plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of x values
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_exp = np.exp(x)

plt.figure(figsize=(10, 6))

plt.plot(x, y_sin, color='blue', linestyle='-', label='sin(x)')
plt.plot(x, y_cos, color='orange', linestyle='--', label='cos(x)')
plt.plot(x, y_exp, color='green', linestyle=':', label='exp(x)')
plt.title('Plot of sin(x), cos(x), and exp(x)', fontsize=14)

plt.xlabel('x', fontsize=12)
plt.ylabel('Function values', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='#FF5733', edgecolor='black', alpha=0.7)

plt.title('Histogram of 1000 Random Numbers from a Normal Distribution', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Sample data
np.random.seed(42)
x_line = np.linspace(0, 10, 50)
y_line = np.sin(x_line)
categories = ['A', 'B', 'C', 'D']
values = [15, 25, 30, 20]
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
data_hist = np.random.normal(loc=0, scale=1, size=1000)

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].plot(x_line, y_line, color='blue')
axs[0, 0].set_title('Line Plot')
axs[0, 0].grid(True)

axs[0, 1].bar(categories, values, color='green')
axs[0, 1].set_title('Bar Chart')

axs[1, 0].scatter(x_scatter, y_scatter, color='red')
axs[1, 0].set_title('Scatter Plot')

axs[1, 1].hist(data_hist, bins=30, color='purple', alpha=0.7)
axs[1, 1].set_title('Histogram')

fig.suptitle('2×2 Grid of Different Plot Types', fontsize=16, y=1.02)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

months = np.arange(1, 13)

# data for three products
product_A = np.array([100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650])
product_B = np.array([80, 120, 160, 200, 240, 280, 320, 360, 400, 440, 480, 520])
product_C = np.array([60, 90, 120, 150, 180, 210, 240, 270, 300, 330, 360, 390])

plt.figure(figsize=(10, 6))
plt.stackplot(months, product_A, product_B, product_C, labels=['Product A', 'Product B', 'Product C'], colors=['#FF9999', '#66B3FF', '#99FF99'])

plt.title('Cumulative Sales of Products Over 12 Months', fontsize=14)
plt.xlabel('Months', fontsize=12)
plt.ylabel('Cumulative Sales', fontsize=12)

plt.legend(loc='upper left')

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data= Student exam scores for three subjects
data = {
    'Math': [88, 92, 79, 85, 90, 95, 70, 80, 75, 85],
    'Science': [78, 82, 88, 90, 85, 87, 92, 75, 80, 84],
    'English': [85, 87, 90, 92, 88, 91, 86, 84, 89, 93]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
boxplot = df.boxplot(column=['Math', 'Science', 'English'], grid=False)

plt.title('Boxplot of Student Exam Scores', fontsize=14)
plt.ylabel('Scores', fontsize=12)
plt.xlabel('Subjects', fontsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data: Market share of four companies
labels = ['Company A', 'Company B', 'Company C', 'Company D']
sizes = [40, 30, 20, 10]  # Market shares in percentage
colors = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99']  # Colors for each segment


explode = (0.1, 0, 0, 0)

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

plt.axis('equal')

plt.title('Market Share of Companies', fontsize=14)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# data
np.random.seed(42)  # For reproducibility
x = np.linspace(0, 10, 100)
y = np.random.rand(100)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Random Data', color='blue')

plt.gca().invert_xaxis()

plt.legend(loc='upper left', fontsize=12, bbox_to_anchor=(1, 1))
plt.title('Random Line Plot with Customized Axis and Legend', fontsize=14)
plt.xlabel('X-axis (Reversed)', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)

plt.grid(True)

plt.tight_layout()
plt.show()
```
